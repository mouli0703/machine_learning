```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('kc_house_data.csv')
data
```

```python
data.columns
```

```python
data = data[['price', 'waterfront', 'sqft_living', 'bedrooms',
             'bathrooms', 'floors', 'condition', 'grade']]
data
```

```python
data.loc[10:20,'bedrooms'] = np.nan
```

```python
data.head(30)
```

```python
missing = data.isna().sum()
```

```python
sns.barplot(x=missing.index,y=missing.values)
plt.xticks(rotation = 45);
plt.show()
```

```python
sns.histplot(data['price'],kde=True,bins=30)
plt.show()
```

```python
sns.heatmap(data.corr(numeric_only=True),annot=True,fmt='.2f',cmap='coolwarm')
plt.show()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(strategy='median')
data['bedrooms'] = imp.fit_transform(data[['bedrooms']])
```

```python
data.head(30)
```

```python
data['waterfront'].dtype
```

```python
data.loc[50,'price']=data['price'].mean()*10
```

```python
sns.boxplot(x=data['price'],palette='bright')
plt.show()
```

```python
q1=data['price'].quantile(0.25)
q3=data['price'].quantile(0.75)
iqr = q3-q1
lower = q1-1.5*iqr
upper = q3+1.5*iqr
```

```python
upper
```

```python
data = data[(data['price']>=lower)&(data['price']<=upper)]
```

```python
data['price'].min()
```

```python
data['price'].max()
```

```python
sns.boxplot(x=data['price'],palette='bright')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
SSC = StandardScaler()
```

```python
data[['sqft_living','bedrooms','bathrooms',
                  'floors','grade','condition']]=SSC.fit_transform(data[['sqft_living','bedrooms','bathrooms',
                  'floors','grade','condition']])

data
```

```python
X=data.drop('waterfront',axis=1)
y=data['waterfront']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
```

```python
knn_cl = KNeighborsClassifier(n_neighbors=5,metric = 'l2')
```

```python
knn_cl.fit(X_train,y_train)
```

```python
y_pred = knn_cl.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

# regressor

```python
X = data.drop('price',axis=1)
y=data['price']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
knn_rg = KNeighborsRegressor(n_neighbors=5,metric = 'l2')
```

```python
knn_rg.fit(X_train,y_train)
```

```python
y_pred = knn_rg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
